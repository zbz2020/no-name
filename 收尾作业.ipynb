```python
import PIL.Image
from calendar import isleap
import datetime
import time
import matplotlib.pyplot as plt 
```

```python
#获取今日日期，判断是今年第几天
day_time=time.strftime('%Y-%m-%d',time.localtime(time.time()))
year=int(day_time[0:4])
month=int(day_time[5:7])
day=int(day_time[8:10])
months = (0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334)
if 0 < month <= 12:
    sum = months[month - 1]
else:
    print('填写的日期错误')
sum += day
leap = 0
if (year % 400 == 0) or ((year % 4 == 0) and (year % 100 != 0)):
    leap = 1
if (leap == 1) and (month > 2):
    sum += 1
print('这是今年的第%d天'%sum)
```

```python
img=PIL.Image.open('test.png')
angle=(sum/360)*365
print('今年已经旋转的角度为%s'%angle)
img= img.rotate(angle) 
plt.imshow(img)
```
